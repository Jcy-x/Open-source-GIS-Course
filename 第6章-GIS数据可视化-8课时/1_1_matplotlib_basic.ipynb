### 介绍matplotlib绘图基础
介绍画布figure, 子图axes, 子图坐标系Axis，子图标题title等。


```python
import matplotlib.pyplot as plt
from osgeo import gdal
```

```python
path_img = 'data/Section-5/s2_chenggong_mndwi_thre025.tif'
```

```python
dset_mndwi = gdal.Open(path_img)
img_array = dset_mndwi.ReadAsArray()
print(img_array.shape)
```

```python
### 在画布上画子图，默认子图数量为1
fig = plt.figure()   ## 画布
plt.imshow(img_array)
```

```python
### 自定义子图（方法2:add_subplot()）
fig = plt.figure(figsize=(10, 4))
axe_1 = fig.add_subplot(1,2,1)   ###以上两行代码可用如下替换：fig, ax = plt.subplots(1,2, figsize=(10,4))
axe_1.set_xlabel('column')
axe_1.set_ylabel('row')
axe_1.set_title('mndwi image of chenggong')
axe_1.imshow(img_array)
axe_2 = fig.add_subplot(1,2,2)
axe_2.imshow(img_array)
```
